```python
# Problem 2 - The logistic map
# the map x->lambda*x(1-x) is called the logistic map

# lambda=1
solve(x*(1-x)-x,x)
# we see that it has only one fixed point which is 0
```

```python
# we set the initial value
x=0.5
x
```

```python
# compute the first 100 terms of the sequence x(k+1)=x(k)*(1-x(k))
# with the mention that x(0)=0.5, as stated before
f=[0 for a in range(101)]

for i in range(1,100):
    x=x*(1-x)
    f[i]=x

# we save all these values in a list with 100 components
```

```python
line([(i,f[i]) for i in range(1,100)])
# in the graph below there are represented 100 values of the sequence, seen on the vertical
# the sequence below seems to monotonically decrease to 0
```

```python
# we set the initial value
x=0.95
x
```

```python
# Problem 3 - Sensitivity with respect to small errors is a feature of chaos
x0=0.67
x1=0.67001
# set the initial values
```

```python
for i in range(1,40):
    x0=4*x0*(1-x0)
    x1=4*x1*(1-x1)
    
print(x0, x1)
# this happens because of 
# the map x->4*x*(1-x) that is chaotic
# and the round-off errors are different
```
