```python
x, y = var('x y')
f = 2*x-x**2-x*y
g = -y+x*y
solve([f==0, g==0], x, y)
```

```python
Jm1 = jacobian([f, g], (x,y))
Jm1
```

```python
A1 = Jm1.subs(x=0, y=0)
A1.eigenvalues()
```

```python
A2 = Jm1.subs(x=2, y=0)
A2.eigenvalues()
```

```python
A3 = Jm1.subs(x=1, y=1)
A3.eigenvalues()
```
